# Matplotlib

본 선형대수학 그래프는 전부 파이썬의 Matplotlib 라이브러리를 사용하여 시각화해서 표현한다.

## 1.1 기본 2차원 벡터

```python
import math
import numpy as np
import matplotlib.pyplot as plt

#벡터 생성(x값, y값)
a = [4, 12]
b = [16, 9]
```

```python
plt.figure(figsize = (7,7))        #그래프를 보여주는 창크기
plt.xlim(0, 20)                    #그래프 최소,최대 x값
plt.ylim(0, 20)                    #그래프 최소,최대 y값
plt.xticks(ticks = np.arange(0, 20, step = 1))   #x축 눈금 표시
plt.yticks(ticks = np.arange(0, 20, step = 1))   #y축 눈금 표시

#벡터 그리기
#(시작점x, 변화율dx, 화살표 높이, 화살표 길이, 색상)
plt.arrow(0,0,a[0], a[1], head_width=0.5, head_length=0.5, color='red')
plt.arrow(0,0,b[0], b[1], head_width=0.5, head_length=0.5, color='green')

#수직선&수평선 그리기(축) - 중심을 나타낼 때 사용한다.
plt.axhline(0, color = 'blue', alpha = 1)
plt.axvline(0, color = 'blue', alpha = 1)

plt.title('Vector')      #제목
plt.grid()               #눈금 표시
plt.show()
```
