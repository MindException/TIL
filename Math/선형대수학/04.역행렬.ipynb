# 역행렬

역행렬의 제일 중요한 성질은 행렬과 역행렬을 곱하였을 경우에 단위행렬이 되는 것이다.

$$ AA^{-1} = A^{-1}A = I$$

* 따라서 위의 성질 중에 $AA^{-1} = A^{-1}A$ 가 성립하지 않는다면 역행렬을 가질 수 없다.

# 기본 행렬(_elementary matrix_)

단위 행렬에 아래와 같이 3가지 행위를 한 경우 기본 행렬이라고 부른다.
1. 두 선형방정식의 위치를 교환하는 것
2. 선형방정식의 양변에 0이 아닌 상수를 곱하는 것
3. 특정 선형방정식에 0이 아닌 상수배를 다른 선형방정식에 더하는 것

## 예시

$$ I_{3} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & 1
\end{pmatrix}
 $$ 

### 1. 두 선형방정식의 위치를 교환하는 것

$$ E_{1} =  
\begin{pmatrix}
  0 & 1 & 0 \\
  1 & 0 & 0 \\
  0 & 0 & 1
\end{pmatrix}
 $$ 
 
### 2. 선형방정식의 양변에 0이 아닌 상수를 곱하는 것

$$ E_{2} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & 3
\end{pmatrix}
 $$ 
 
### 3. 특정 선형방정식에 0이 아닌 상수배를 다른 선형방정식에 더하는 것

$$ E_{3} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  -3 & 0 & 1
\end{pmatrix}
 $$ 

## 기본행렬의 역행렬 

기본행렬은 가역행렬이며, 기본 행렬의 역행렬은 기본행렬이다.

### 1. 두 선형방정식의 위치를 교환하는 것

$$ E_{1} =  
\begin{pmatrix}
  0 & 1 & 0 \\
  1 & 0 & 0 \\
  0 & 0 & 1
\end{pmatrix},\qquad
E_{1}^{-1} =  
\begin{pmatrix}
  0 & 1 & 0 \\
  1 & 0 & 0 \\
  0 & 0 & 1
\end{pmatrix}
 $$ 
 
### 2. 선형방정식의 양변에 0이 아닌 상수를 곱하는 것

$$ E_{2} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & x
\end{pmatrix},\qquad
E_{2}^{-1} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & \frac{1}{x}
\end{pmatrix}
 $$ 
 
### 3. 특정 선형방정식에 0이 아닌 상수배를 다른 선형방정식에 더하는 것

$$ E_{3} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  -x & 0 & 1
\end{pmatrix},\qquad
E_{3}^{-1} =  
\begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  x & 0 & 1
\end{pmatrix}
 $$ 

## 기본행렬 곱을 통한 역행렬 계산

__행동치__ 를 이용하여 $A$가 $I$로 될 수 있다면 역행렬을 구할수 있다는 개념이다.
* 정방행렬인 A의 앞에 곱해서 만드며

$$
E_{n} \cdots E_{2}E_{1}A = I 
$$ 

$$
A^{-1} = E_{n} \cdots E_{2}E_{1}
$$
를 의미한다

* 이렇게 역행렬을 구하는 것이 컴퓨터로 매우 느려서 나온 것이 LU분해이다.

### 행동치(_row equivalence_)란?

* 행렬 A와 B에 대하여, A를 기본행렬을 이용하여 B로 바꿀 수 있는 것을 의미하며 __A ~ B__ 로 표기한다.

$$
B = E_{n} \cdots E_{2}E_{1}A
$$

* 행동치가 중요한 이유는 계산 도 중 0이 나오면 역행렬이 존재하지 않는 것을 알 수 있다.

## 문제) 역행렬을 구하여라

### (a) $ \begin{pmatrix}
  3 & 4 \\
  1 & 2 
\end{pmatrix}^{-1} = \; ?
$

$$
\begin{bmatrix}
1&0&|&1&-2\\
1&2&|&0&1\\ 
\end{bmatrix}, 
\begin{bmatrix}
1&0&|&1&-2\\
0&2&|&-1&3\\ 
\end{bmatrix},
\begin{bmatrix}
1&0&|&1&-2\\
0&1&|&\frac{-1}{2}&\frac{3}{2}\\ 
\end{bmatrix}
$$

$$
\begin{pmatrix}
  3 & 4 \\
  1 & 2 
\end{pmatrix}^{-1} = 
\begin{pmatrix}
1&-2\\
\frac{-1}{2}&\frac{3}{2}\\ 
\end{pmatrix}
$$

### (B) $ \begin{pmatrix}
  1 & 0 & 5\\
  1 & 1 & 1\\
  0 & 1 & -4
\end{pmatrix}^{-1} = \; ?
$

$$
\begin{bmatrix}
  1 & 0 & 5 &|& 1 & 0 & 0\\
  1 & 1 & 1 &|& 0 & 1 & 1\\
  0 & 1 & -4 &|& 0 & 0 & 1
\end{bmatrix}, \qquad
\begin{bmatrix}
  1 & 1 & 1 &|& 1 & 0 & 1\\
  1 & 1 & 1 &|& 0 & 1 & 0\\
  0 & 1 & -4 &|& 0 & 0 & 1
\end{bmatrix}
$$

* 행 동치에서 000이 존재하게 됨으로 역행렬이 존재하지 않는다.

# 동차 연립 방정식 특징

$Ax = 0$의 향태인 연립 방정식을 동차 연립방정식이라고 한다.

1. $x = 0$인 해를 자명해(_trivial solution_)
2. $x \neq 0$인 해를 비자명해(_non-trivial solution_)
3. $Ax = 0$에서 A가 가역이면 자명해만 존재한다.
4. 미지수의 개수가 선형 방정식의 개수보다 많은 동차 연립선형방정식 $Ax = 0$는 비자명해를 갖는다.
    * 행의 초과한 열의 미지수는 자유변수가 되어서 해를 무수히 많이 갖게 된다.
